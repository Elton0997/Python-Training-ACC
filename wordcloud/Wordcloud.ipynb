```python
with open("words1302.txt", 'r') as fh:  
    filedata = fh.read()

print("File data sample : ", filedata[:200])


from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)


stopwords.update(["many","using", "want", "value", "Test"])
wordcloud = WordCloud(stopwords=stopwords, max_words=50, \
                      background_color="white").generate(filedata)

```

```python
import matplotlib.pyplot as mpLib
mpLib.imshow(wordcloud)
mpLib.axis("off")
mpLib.show()
```

```python
stopwords.update(["many","using", "want", "value"])

wordcloud = WordCloud(stopwords=stopwords, max_words=100, \
                      background_color="azure").generate(filedata)

mpLib.imshow(wordcloud)
mpLib.axis("off")
mpLib.show()
```
